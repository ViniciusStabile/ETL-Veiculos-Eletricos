```python
df = df[df["County"].notnull()]
```

```python
df[df["Electric Range"].isnull()][["Model Year","Model"]]
```

```python
electric_range_null = df[df['Electric Range'].isnull()]['Model'].unique().tolist()
electric_range_null
```

```python
df[df['Model'].isin(electric_range_null) & (df["Model Year"] == 2025)][['Model Year','Model','Electric Range']]
```

```python
electric_range_null = df[df['Electric Range'].isnull()]['Model'].unique().tolist()
electric_range_null
```

```python
df[df['Model'].isin(electric_range_null) & (df["Model Year"] == 2025)][['Model Year','Model','Electric Range']]
```

```python
# Preenchendo valores NaN somente nas linhas em que Model é 'GLC-CLASS' e Model Year é 2025
df.loc[(df['Model'] == 'GLC-CLASS') & (df['Model Year'] == 2025), ['Electric Range']] = df.loc[(df['Model'] == 'GLC-CLASS') & (df['Model Year'] == 2025), ['Electric Range']].fillna(54.0)
df.loc[(df['Model'] == 'GLC-CLASS') & (df['Model Year'] == 2025), ['Base MSRP']] = df.loc[(df['Model'] == 'GLC-CLASS') & (df['Model Year'] == 2025), ['Base MSRP']].fillna(0)
```

```python
df.isnull().sum()
```

```python
df[df['Electric Range'].isnull()][['Model','Model Year','Electric Range','Base MSRP']]
```

```python
df[(df['Model'] == 'PANAMERA') & (df['Model Year'] == 2025)][['Electric Range']]
```

```python
df = df[~((df['Model'] == 'PANAMERA') & (df['Model Year'] == 2025))]
```

```python
df.isnull().sum()
```

```python
df[(df['Model'] == 'CAYENNE') & (df['Model Year'] == 2025)][['Electric Range']]
```

```python
df = df[~((df['Model'] == 'CAYENNE') & (df['Model Year'] == 2025))]
```

```python
df.isnull().sum()
```

```python
df[df['Electric Range'].isnull()][['Model','Model Year','Electric Range','Base MSRP']]
```

```python
df = df[~((df['Model'] == 'SL-CLASS') & (df['Model Year'] == 2025))]
```

```python
df[df['City'] == 'Long Beach'][['County','City','State','Postal Code','Vehicle Location']]
```

```python
df_cleaned = df.dropna(subset=['Vehicle Location'])
```

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned[df_cleaned['Legislative District'].isnull()][['Legislative District', 'County', 'City', 'State']].sort_values(by='County')
```

```python
df_cleaned.dropna(subset=['Legislative District'], inplace=True)
```

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned.reset_index(drop=True, inplace=True)
```

```python
cols_to_clean = ['City', 'County', 'Model','Make','Electric Vehicle Type','Clean Alternative Fuel Vehicle (CAFV) Eligibility','Vehicle Location','Electric Utility']
```

```python
df_cleaned[cols_to_clean] = df_cleaned[cols_to_clean].apply(lambda x: x.str.upper())
```

```python
import boto3

from google.colab import userdata

aws_access_key_id = userdata.get('AWS_ACCESS_KEY_ID')
aws_secret_access_key = userdata.get('AWS_SECRET_ACCESS_KEY')

boto3.setup_default_session(
    aws_access_key_id=aws_access_key_id,
    aws_secret_access_key=aws_secret_access_key,
    region_name="sa-east-1"
)

s3 = boto3.client("s3")
```

```python
from io import BytesIO

parquet_buffer = BytesIO()
df_cleaned.to_parquet(parquet_buffer, index=False)


s3.put_object(
     Bucket="vinicius-data-pipeline",
     Key=f"bronze/dados_carros_eletricos.parquet",
     Body=parquet_buffer.getvalue()
 )
```

```python
response = s3.list_objects(
    Bucket="vinicius-data-pipeline",
    Prefix="bronze/"
)
```
